```
north = []
south = []
west = []

for f in sorted(os.listdir("maurons")):
    file = str(nbt.read_from_nbt_file(os.path.join("maurons", f)))
    f = f.split(".")[0].split("_")[2]
    if "north" in file:
        north.append(f)
    elif "south" in file:
        south.append(f)
    elif "west" in file:
        west.append(f)

res = {
    "north": north,
    "south": south,
    "west": west
}

print(json.dumps(res, indent=4))

del north
del south
del west
del res
```
