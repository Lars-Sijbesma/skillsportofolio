```
# Data
generaties = [1, 2, 3, 4, 5, 6, 7]
aanvallen = [11254, 6996, 10420, 8846, 12567, 5510, 6771]

# Haal Set1 kleuren op
kleuren = sns.color_palette("Set1", n_colors=len(generaties))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Normale schaal
ax1.bar(generaties, aanvallen, color=kleuren)
ax1.set_title('Normale schaal')
ax1.set_xlabel('Generatie')
ax1.set_ylabel('Gemiddelde aanval')
ax1.set_ylim(0, max(aanvallen) + 1000)
for i, v in enumerate(aanvallen):
    ax1.text(generaties[i] - 0.25, v + 150, str(v), color='black')

# Gemanipuleerde schaal
ax2.bar(generaties, aanvallen, color=kleuren)
ax2.set_title('Misleidende schaal')
ax2.set_xlabel('Generatie')
ax2.set_ylim(6400, 13000)
for i, v in enumerate(aanvallen):
    ax2.text(generaties[i] - 0.25, v + 100, str(v), color='black')

plt.tight_layout()
plt.show()
```
