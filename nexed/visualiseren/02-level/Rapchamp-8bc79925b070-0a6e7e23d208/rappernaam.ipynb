```python
import random
```

```python
naam = input("Wat is jouw naam? ")
meubelstuk = input("Wat is jouw favorieten meubelstuk ")
keukengerei = input("Welk voorwerp uit de keuken heb je het laatst gebruikt? ")
voedsel = input("Wat is je lievelingseten? ")
kleur = input("Wat is jouw lievelingskleur? ")
```

```python
def genereer_rapper_naam(naam, kleur, gek_woord):
    voorstukje = random.choice([naam, kleur, gek_woord, meubelstuk, keukengerei, voedsel])

    if random.random() < 0.25:
        return voorstukje + ' " ' + naam + ' " ' + gek_woord
    else:
        return voorstukje + ' ' + gek_woord
```

```python
gek_woord = random.choice([meubelstuk, keukengerei, voedsel])
rappernaam = genereer_rapper_naam(naam, kleur, gek_woord)
print("Jouw naam is " + rappernaam)
```
