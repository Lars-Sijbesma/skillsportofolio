```


# Data inlezen
df = pd.read_csv('mc_menu_final.csv')

# Filter op categorie "Smoothies & Shakes"
shakes_df = df[df['Category'] == 'Smoothies & Shakes']

# Barchart maken
plt.figure(figsize=(10, 6))  # Grotere figuur voor betere leesbaarheid
plt.bar(shakes_df['Item'], shakes_df['Total Fat'])

# Labels en titel
plt.xlabel('Item')
plt.ylabel('Total Fat')
plt.title('Total Fat in Smoothies & Shakes')

# Draai x-labels schuin zodat ze niet overlappen
plt.xticks(rotation=45, ha='right')

# Layout verbeteren
plt.tight_layout()

# Plot tonen
plt.show()
```
