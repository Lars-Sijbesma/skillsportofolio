```
weather = pd.read_csv('weather_netherlands.csv', index_col=0, parse_dates=True)
```

```
weather.plot()
plt.show()
```

```
weather["AverageTemperature"].plot()
plt.show()
```

```
rotterdam_weather = weather[weather['City'] == 'Rotterdam']
max_temp = rotterdam_weather['AverageTemperature'].max()
print(f"{max_temp:.3f}°C")
```

```
maastricht_weather = weather[weather['City'] == 'Maastricht']
min_temp = maastricht_weather['AverageTemperature'].min()
print(f"{min_temp:.2f}°C")
```

```
aantal_unieken_steden = weather['City'].nunique()
print(f"{aantal_unieken_steden:}")
```

```
voorkomende_weather = weather['AverageTemperature'].mode()
print(f"{floor(voorkomende_weather[len(voorkomende_weather)-1])}°C")
```

```
top_3 = weather['AverageTemperature'].nlargest(3)
print(f"{top_3}")
```
