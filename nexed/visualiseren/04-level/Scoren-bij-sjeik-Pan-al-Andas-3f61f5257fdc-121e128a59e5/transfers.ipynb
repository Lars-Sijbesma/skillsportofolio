```
import pandas as pd
import matplotlib.pyplot as plt
```

```
fifa_2 = pd.read_csv('fifa_ratings.csv')
print(fifa_2)
```

tellen

```
club_values = fifa_2.groupby('club')['value_eur'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(club_values.index, club_values.values / 1000000)
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Most Valuable Football Clubs')
plt.xlabel('Club')
plt.ylabel('Total Player Value (Million EUR)')
plt.tight_layout()
plt.show()
```
